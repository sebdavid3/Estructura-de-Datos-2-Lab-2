```python
import collections
collections.Iterable = collections.abc.Iterable
import haversine as hs
from haversine import Unit
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import folium
from folium.plugins import MarkerCluster
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab/flights_final.csv')
df
```

```python
def Flight_Distance(i):
  loc1=(df.loc[i,'Source Airport Latitude'], df.loc[i,'Source Airport Longitude'])
  loc2=(df.loc[i,'Destination Airport Latitude'],df.loc[i,'Destination Airport Longitude'])
  result=hs.haversine(loc1,loc2,unit=Unit.KILOMETERS)
  return result

distancia=[]
for index in df.index:
  distancia.append(Flight_Distance(index))

df =df.join(pd.DataFrame(distancia, columns=['Flight Distance (Km)']))
df.head()
```

```python
vuelos = df.drop_duplicates()
vuelos = vuelos.reset_index(drop=True)
vuelos
```

```python
class Airport:
    def __init__(self, code, name, city, country, latitude, longitude):
        self.code = code
        self.name = name
        self.city = city
        self.country = country
        self.latitude = latitude
        self.longitude = longitude

def create_graph(df):
        F = nx.Graph()
        for index, row in df.iterrows():
            aeropuerto_d = Airport(
                row['Destination Airport Code'],
                row['Destination Airport Name'],
                row['Destination Airport City'],
                row['Destination Airport Country'],
                row['Destination Airport Latitude'],
                row['Destination Airport Longitude'],
            )
            F.add_node(aeropuerto_d.code, airport=aeropuerto_d)

            aeropuerto_origen = Airport(
                row['Source Airport Code'],
                row['Source Airport Name'],
                row['Source Airport City'],
                row['Source Airport Country'],
                row['Source Airport Latitude'],
                row['Source Airport Longitude'],
            )

            if aeropuerto_origen.code not in F.nodes:
                F.add_node(aeropuerto_origen.code, airport=aeropuerto_origen)

            F.add_edge(aeropuerto_origen.code, aeropuerto_d.code, weight=row['Flight Distance (Km)'])

        return F

# Muestra informacion del aereopuerto apartir de su codigo
def show_node_information(graph, code):
        if code not in graph.nodes:
            print(f"Error: El aereopuerto con código {code} no existe en el grafo.")
            return

        airport = graph.nodes[code]['airport']
        print(f"Datos del codigo ({code}):")
        for key, value in airport.__dict__.items():
            print(f"\t{key}: {value}")


# Se obtiene la distancia de una conexion
def get_edge_weight(graph, source, destination):
    if not graph.has_edge(source, destination):
        return None
    return graph[source][destination]['weight']

# Muestra las conexiones de un aereopuerto
def show_connected_nodes(graph, node_code):

  if node_code not in graph.nodes:
    print(f"El codigo {node_code} no existe")
    return

  connected_nodes = list(graph.neighbors(node_code))

  if not connected_nodes:
    print(f"El aereopuerto {node_code} no tiene conexiones.")
    return
  else:
    print(f"Conecciones de {node_code}:")
    for connected_node in connected_nodes:
      edge_weight = get_edge_weight(graph, node_code, connected_node)
      connected_airport = graph.nodes[connected_node]["airport"]
      print(f"\t- {connected_airport.code} ({connected_airport.name})", end="")
      if edge_weight is not None:
        print(f" (distancia: {edge_weight} Km)")
      else:
        print()

# Calcula el peso total apartir de un camino
def calculate_total_weight(graph, path):
    total_weight = 0
    for i in range(len(path) - 1):
        weight = get_edge_weight(graph, path[i], path[i + 1])
        if weight is not None:
            total_weight += weight
    return total_weight

def shortest_path_to_all_airports(graph, origin):
    all_airports = list(graph.nodes)
    shortest_paths = {}
    for airport in all_airports:
        try:
            shortest_path = nx.dijkstra_path(graph, origin, airport)
            total_distance = calculate_total_weight(graph, shortest_path)
            shortest_paths[airport] = (shortest_path, total_distance)
        except nx.NetworkXNoPath:
            pass
    return shortest_paths


def print_longest_shortest_flights(graph,code):
  shortest_paths = shortest_path_to_all_airports(graph,code)
  sorted_airports = sorted(shortest_paths.items(), key=lambda item: item[1][1], reverse=True)

  for i, (airport, (path, distance)) in enumerate(sorted_airports):
      if i < 10:
          print(f"Camino mas corto a {airport} ({', '.join(path)}):")
          print(f"\tDistancia Total: {distance} Km")



def print_shortest_path(graph, origin, destination):

    try:
        shortest_path = nx.dijkstra_path(graph, origin, destination)

        if not shortest_path:
            print(f"No hay camino entre {origin} y {destination}.")
            return

        total_distance = calculate_total_weight(graph, shortest_path)

        origin_airport = graph.nodes[origin]['airport']
        map = folium.Map(location=[origin_airport.latitude, origin_airport.longitude])

        marker_cluster = MarkerCluster()
        for airport_code in shortest_path:
            airport = graph.nodes[airport_code]['airport']
            marker = folium.Marker([airport.latitude, airport.longitude],
                                   popup=f"\n{airport.name}\n{airport.code}\n{airport.city}\n{airport.country}")
            map.add_child(marker)



        folium.PolyLine(
            [
                (graph.nodes[code]['airport'].latitude, graph.nodes[code]['airport'].longitude)
                for code in shortest_path
            ],
            color="red",
            weight=2.5,
            opacity=1,
            tooltip= folium.Tooltip(f"Distancia ({total_distance} Km)"),
        ).add_to(map)
        map.save("Camino_minimo.html")

        print(f"El mapa se ha generado: camino_minimo.html")

    except nx.NetworkXNoPath:
        print(f"Error")

        return map
```
