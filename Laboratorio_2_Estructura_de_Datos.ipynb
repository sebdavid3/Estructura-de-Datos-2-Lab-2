```python
import collections
collections.Iterable = collections.abc.Iterable
import haversine as hs
from haversine import Unit
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import folium
from folium.plugins import MarkerCluster
```

```python
def Flight_Distance(i):
  loc1=(df.loc[i,'Source Airport Latitude'], df.loc[i,'Source Airport Longitude'])
  loc2=(df.loc[i,'Destination Airport Latitude'],df.loc[i,'Destination Airport Longitude'])
  result=hs.haversine(loc1,loc2,unit=Unit.KILOMETERS)
  return result

distancia=[]
for index in df.index:
  distancia.append(Flight_Distance(index))
df
```

```python
class Airport:
    def __init__(self, code, name, city, country, latitude, longitude):
        self.code = code
        self.name = name
        self.city = city
        self.country = country
        self.latitude = latitude
        self.longitude = longitude

def create_graph(df):
        F = nx.Graph()
        for index, row in df.iterrows():
            aeropuerto_d = Airport(
                row['Destination Airport Code'],
                row['Destination Airport Name'],
                row['Destination Airport City'],
                row['Destination Airport Country'],
                row['Destination Airport Latitude'],
                row['Destination Airport Longitude'],
            )
            F.add_node(aeropuerto_d.code, airport=aeropuerto_d)

            aeropuerto_origen = Airport(
                row['Source Airport Code'],
                row['Source Airport Name'],
                row['Source Airport City'],
                row['Source Airport Country'],
                row['Source Airport Latitude'],
                row['Source Airport Longitude'],
            )

            if aeropuerto_origen.code not in F.nodes:
                F.add_node(aeropuerto_origen.code, airport=aeropuerto_origen)

            F.add_edge(aeropuerto_origen.code, aeropuerto_d.code, weight=row['Flight Distance (Km)'])

        return F

def show_node_information(graph, code):
        if code not in graph.nodes:
            print(f"Error: El nodo con código {code} no existe en el grafo.")
            return

        airport = graph.nodes[code]['airport']
        print(f"Datos del nodo ({code}):")
        for key, value in airport.__dict__.items():
            print(f"\t{key}: {value}")



def get_edge_weight(graph, source, destination):
    if not graph.has_edge(source, destination):
        return None
    return graph[source][destination]['weight']


def show_connected_nodes(graph, node_code):


  if node_code not in graph.nodes:
    print(f"Error: The node with code {node_code} does not exist in the graph.")
    return

  connected_nodes = list(graph.neighbors(node_code))

  if not connected_nodes:
    print(f"Node {node_code} has no connections (destinations) in the graph.")
  else:
    print(f"Connections (destinations) of node {node_code}:")
    for connected_node in connected_nodes:
      edge_weight = get_edge_weight(graph, node_code, connected_node)
      connected_airport = graph.nodes[connected_node]["airport"]
      print(f"\t- {connected_airport.code} ({connected_airport.name})", end="")
      if edge_weight is not None:
        print(f" (distance: {edge_weight} Km)")
      else:
        print()

def longest_airports(graph, code, max_results=10):
  dijkstra_result = nx.single_source_dijkstra_path_length(graph, code)
  sorted_results = sorted(dijkstra_result.items(), key=lambda x: x[1], reverse=True)

  longest_airports = [airport for airport, distance in sorted_results[:max_results]]
  return longest_airports


def longest_airports_with_paths(graph, code, max_results=10):

    try:
        dijkstra_result = nx.single_source_dijkstra_path_length(graph, code)
    except nx.NetworkXNoPath:
        print(f"No paths found from airport {code}.")
        return {}

    # Sort airports and distances in descending order (longest first)
    sorted_results = sorted(dijkstra_result.items(), key=lambda x: x[1], reverse=True)

    longest_airports_with_paths = {}
    for i, (airport, distance) in enumerate(sorted_results):
        if i >= max_results:
            break

        try:
            longest_path = nx.shortest_path(graph, code, airport, method='dijkstra')
        except nx.NetworkXNoPath:
            print(f"No path found to airport {airport} from {code}.")
            continue

        longest_airports_with_paths[airport] = longest_path

    return longest_airports_with_paths

def calculate_total_weight(graph, path):
    total_weight = 0
    for i in range(len(path) - 1):
        weight = get_edge_weight(graph, path[i], path[i + 1])
        if weight is not None:
            total_weight += weight
    return total_weight

def print_longest_flights(graph, code):

  longest_airports_data = longest_airports_with_paths(graph, code, max_results=10)

  if longest_airports_data:
    sorted_airports = sorted(longest_airports_data.items(), key=lambda item: calculate_total_weight(graph, item[1]), reverse=True)

    for airport, path in sorted_airports:
      total_weight = calculate_total_weight(graph, path)
      print(f"Longest path to {airport} ({', '.join(path)}):")
      print(f"\tTotal distance: {total_weight} Km")

      for airport_code in path:
        show_node_information(graph, airport_code)

      print("------------------------------------------------------------------------------------------")
  else:
    print("No aereopuertos encontrados")

def print_shortest_path(graph, origin, destination):

  try:
    shortest_path = nx.shortest_path(graph, origin, destination, method='dijkstra')

    if not shortest_path:
      print(f"No camino entre {origin} y {destination}.")
      return

    total_distance = calculate_total_weight(graph, shortest_path)

    print(f"Camino mas corto entre {origin} a {destination}:")
    print(f"\tCamino: {' - '.join(shortest_path)}")
    print(f"\tDistancia Total: {total_distance} Km")

    for airport_code in shortest_path:
      show_node_information(graph, airport_code)

  except nx.NetworkXNoPath:
    print(f"Error")


def print_shortest_path(graph, origin, destination):

    try:
        shortest_path = nx.shortest_path(graph, origin, destination, method='dijkstra')

        if not shortest_path:
            print(f"No hay camino entre {origin} y {destination}.")
            return

        total_distance = calculate_total_weight(graph, shortest_path)

        origin_airport = graph.nodes[origin]['airport']
        map = folium.Map(location=[origin_airport.latitude, origin_airport.longitude], zoom_start=6)

        marker_cluster = MarkerCluster()
        for airport_code in shortest_path:
            airport = graph.nodes[airport_code]['airport']
            marker = folium.Marker([airport.latitude, airport.longitude], popup=f"{airport.name} ({airport.code})")
            marker_cluster.add_child(marker)

        map.add_child(marker_cluster)

        # Create a tooltip for the path
        tooltip = folium.Tooltip(f"Shortest Path ({total_distance} Km)")

        folium.PolyLine(
            [
                (graph.nodes[code]['airport'].latitude, graph.nodes[code]['airport'].longitude)
                for code in shortest_path
            ],
            color="blue",
            weight=2.5,
            opacity=1,
            tooltip=tooltip,
        ).add_to(map)

        map.save("Camino_minimo.html")

        print(f"El mapa se ha generado: camino_minimo.html")

    except nx.NetworkXNoPath:
        print(f"Error")
```
